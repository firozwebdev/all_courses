```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a NumPy array with sample data
x = np.arange(11)
y1 = x + 2
y2 = x + 3

# Create a Pandas DataFrame with the NumPy arrays
df = pd.DataFrame({'x': x, 'x+2': y1, 'x+3': y2})

# Plot the line chart
df.plot.line(x='x', y=['x+2', 'x+3'],  linestyle='-')


# Add legends
plt.legend( loc='lower right')

# Display the chart
plt.show()
```

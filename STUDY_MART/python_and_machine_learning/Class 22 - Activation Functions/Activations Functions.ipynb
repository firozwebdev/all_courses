The Rectified Linear Unit (ReLU) activation function is a piecewise linear function that will output the input directly if it is positive, otherwise, it will output zero. It has become the default activation function for many types of neural networks because it is easy to train and often achieves better performance.

Here are the steps describing how ReLU works:

1. **Input**: ReLU takes a single number as an input.

2. **Conditional Operation**: It checks if this input is greater than zero.

3. **Output**: If the input is greater than zero, it outputs that number. If the input is less than or equal to zero, it outputs zero.

Mathematically, the ReLU function is defined as:

\[ f(x) = max(0, x) \]

Where `x` is the input to a neuron.

**Usage of ReLU:**

- **Deep Neural Networks (DNNs)**: ReLU is used in almost all the deep networks, as it helps in overcoming the vanishing gradient problem, which is a big issue with sigmoid or tanh functions.
  
- **Convolutional Neural Networks (CNNs)**: ReLU is widely used in CNNs which are the backbone of image recognition, image classification, and other image-related tasks.

- **Generative Adversarial Networks (GANs)**: ReLU is also employed in GANs which are state-of-the-art for generating new content.

- **Other Variants**: There are variants of ReLU like Leaky ReLU and Parametric ReLU that are used to mitigate the issue of dying ReLU (where neurons output zero for all inputs, effectively "dying").

ReLU is popular due to its simplicity and the fact that it reduces training time by accelerating the convergence. It also helps to mitigate the vanishing gradient problem that can occur with traditional activation functions like the sigmoid or tanh functions.

```python
import numpy as np # Algebra
import matplotlib.pyplot as plt # Visualzation

def relu_activation_function(feature):
    return np.maximum(0, feature)

feature = np.linspace(-10, 1, 500)
#print(f"Feature Datasets : {feature}")
predicted_value = relu_activation_function(feature)
#print(f"Comprise Datasets : {predicted_value}")

# Define the Canvas area
plt.figure(figsize=(5, 3))
plt.plot(feature, predicted_value, label = "Relu Activation Function")
plt.title("Relu Activation Function")
plt.xlabel("Feature")
plt.ylabel("Predicted value")
plt.grid(True)
plt.legend()
plt.show()
```

Parametric Rectified Linear Unit (PReLU) is a type of activation function used in neural networks, which is a modification of the traditional Rectified Linear Unit (ReLU). Here's an explanation of PReLU and its implementation step by step:

### What is PReLU?

1. **Rectified Linear Unit (ReLU):** The standard ReLU activation function is defined as \( f(x) = max(0, x) \). It outputs \( x \) if \( x \) is positive and \( 0 \) otherwise. This has been widely used due to its simplicity and efficiency.

2. **Limitation of ReLU:** ReLU can cause what's known as the "dying ReLU" problem, where neurons can sometimes only output zero. This happens when inputs to the neuron are negative, leading to zero gradients during backpropagation which effectively "kills" the neuron, preventing it from learning further.

3. **Parametric ReLU (PReLU):** To overcome this, PReLU introduces a small slope to the negative part of the function, making it \( f(x) = x \) for \( x > 0 \) and \( f(x) = \alpha x \) for \( x \leq 0 \), where \( \alpha \) is a small, learnable parameter. This allows for a small, non-zero gradient when the neuron's input is negative.

### Steps to Implement PReLU:

1. **Initialization:** Initialize the network's weights and biases, along with a small positive value for the parameter \( \alpha \) for each neuron.

2. **Forward Pass:**
   - Calculate the neuron's pre-activation value.
   - Apply PReLU: If the pre-activation value is positive, output it directly. If it's negative, multiply it by \( \alpha \).

3. **Backpropagation:**
   - During the gradient descent step, calculate gradients for both the weights and the \( \alpha \) parameter.
   - Adjust \( \alpha \) along with the weights and biases based on the gradients.

4. **Training:**
   - Continue the process of forward pass and backpropagation for multiple epochs over the training data.
   - Update \( \alpha \) at each step along with other parameters.

5. **Regularization (optional):** Sometimes, regularization techniques might be applied to \( \alpha \) to prevent overfitting, especially if the dataset is small or not diverse.

### Where to Use PReLU:

- **Deep Networks:** PReLU is useful in deep neural networks where the dying ReLU problem is more prevalent.
- **Convolutional Neural Networks (CNNs):** Its often used in CNNs for tasks like image classification and object detection.
- **Problems with Negative Inputs:** Its particularly beneficial in networks where there is an expectation of many negative inputs to neurons.
- **Areas Requiring Robustness:** PReLU helps the network learn more complex patterns and thus can be used in areas requiring robust feature learning.

Using PReLU can help improve model performance by maintaining active gradients and enabling the learning of more complex functions. However, it also introduces extra parameters that need to be learned, which can slightly increase the complexity of the model.

```python
import numpy as np # Algebra
import matplotlib.pyplot as plt # Visualzation

def pRelu_activation_function(feature, alpha):
    return np.where(feature > 0, feature, alpha *feature)

feature = np.linspace(-10, 10, 500)
#print(f"Feature Datasets : {feature}")
alpha = [0.1, 0.2, 0.3, 0.4]

# Define the Canvas area
plt.figure(figsize=(5, 3))
for i_alpha in alpha:
    plt.plot(feature, pRelu_activation_function(feature, i_alpha), label = f"pRelu Activation Function : {i_alpha}")
plt.plot(feature, np.maximum(0, feature), label = "Relu Activation Function", linestyle = "--")
plt.title("pRelu Activation Function")
plt.xlabel("Feature")
plt.ylabel("Predicted value")
plt.grid(True)
plt.legend()
plt.show()
```

The Exponential Linear Unit (ELU) is a type of activation function used in neural networks, which is especially common in the hidden layers of deep neural networks. It aims to solve some of the problems that other activation functions like the Rectified Linear Unit (ReLU) have, such as the dying ReLU problem. Here's a step-by-step explanation of ELU:

### Steps of How ELU Works:

1. **Definition**: ELU is mathematically defined as:

   Here, \( \alpha \) is a hyperparameter that defines the value to which an ELU saturates for negative net inputs.

2. **Positive Inputs**: For positive values of \( x \), ELU behaves exactly like the identity function, so the output is equal to the input.

3. **Negative Inputs**: For negative values of \( x \), ELU outputs values that are greater than or equal to \(-\alpha\), but less than zero. This helps to keep the mean activations closer to zero, which can help speed up learning.

4. **Smoothness**: Unlike ReLU, which abruptly changes slope at \( x = 0 \), the ELU function is smooth everywhere, which helps in optimizing gradient descent as it provides a smoother curve for the derivative.

5. **Differentiability**: ELU is differentiable at all points, which makes it easier to compute gradients during the backpropagation process in training neural networks.

### Where to Use ELU:

1. **Deep Neural Networks**: ELU can be particularly useful in deeper networks where the vanishing gradient problem can be more pronounced, as it helps to mitigate the impact of this problem.

2. **Problems Requiring Fast Convergence**: Since ELU can help the network learn faster (due to mean activations being closer to zero), it can be useful in problems where fast convergence is beneficial.

3. **Avoiding Dead Neurons**: In situations where you might encounter the dying ReLU problem (where neurons stop learning entirely), ELU can be a good alternative as it helps to avoid dead neurons due to its properties for negative input values.

4. **Regression and Classification Tasks**: ELU can be used in both regression and classification tasks within neural networks.

### Considerations:

- **Computationally Intensive**: ELU can be more computationally intensive than ReLU because of the exponential function, so it's essential to consider the trade-off between faster convergence and computational cost.
  
- **Hyperparameter Tuning**: The \( \alpha \) parameter in ELU typically needs to be tuned as a hyperparameter, which may require additional experimentation.

ELU can be a valuable activation function in scenarios where you need a balance between computational efficiency and the need for fast convergence. It's particularly beneficial for maintaining active neurons and promoting faster learning, which makes it suitable for deep learning applications.

```python
import numpy as np # Algebra
import matplotlib.pyplot as plt # Visualzation

def elu_activation_function(feature, alpha=1.0):
    return np.where(feature > 0, feature, alpha *(np.exp(feature) - 1))

feature = np.linspace(-3, 3, 500)
#print(f"Feature Datasets : {feature}")
predicted_value = elu_activation_function(feature)
#print(f"Comprise Datasets : {predicted_value}")

# Define the Canvas area
plt.figure(figsize=(5, 3))
plt.plot(feature, predicted_value, label = "elu Activation Function")
plt.title("elu Activation Function")
plt.xlabel("Feature")
plt.ylabel("Predicted value")
plt.grid(True)
plt.legend()
plt.show()
```

The softmax function, also known as the normalized exponential function, is a function that turns a vector of K real numbers into a vector of K real numbers that sum to 1. The function is often used in machine learning, particularly in the context of neural networks, to map the non-normalized output of a network to a probability distribution over predicted output classes.

Here's a step-by-step explanation of how the softmax function works:

1. **Exponentiate**: Take e (the base of natural logarithms) to the power of each element of the input vector. This step ensures that all numbers are positive because e raised to any real power is always positive.

2. **Normalize**: Sum all the exponentiated values to get a normalization constant. This makes sure that when we divide each exponentiated value by the sum, the results will add up to 1.

3. **Divide**: Divide each exponentiated value by the normalization constant. This step ensures that the resulting vector elements are between 0 and 1 and that their sum is 1.

The softmax function is used in various machine learning algorithms where classification into multiple classes is required, such as:
- **Multinomial Logistic Regression**: It is used at the output layer to provide probabilities that the input belongs to each class.
- **Neural Networks**: It is commonly used at the output layer of a neural network for multi-class classification.

It's important in neural networks because it allows the model's outputs to be interpreted directly as probabilities, which is very useful for classification tasks. Each output corresponds to the model's confidence in the input being of a particular class.
